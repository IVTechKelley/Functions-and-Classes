### Vehicle Data Compiler

Author: Patrick Kelley

Description: A Vehicle and Automobile class prompt the user for car attributes. Then that data is displayed.

Variables - 

type str - The kind of vehicle e.g. car, boat, plane, etc.

year str - The year the car vehicle manufactured.

make str - The brand or company that released the vehicle.

model str - The specific model of vehicle.

doors str - The number of doors on the vehicle. Usually 2 or 4.

roof str - The style of roof. Usually solid or sun roof.

```python
class Vehicle:
    
    # Holds the type of vehicle.
    
    def __init__(self, type):
        self.type = type

    def get_type(self):
        self.type = input("Enter the type of vehicle (Car, Truck, Boat, Plane, Bike, etc.).\n")


class Automobile(Vehicle):
    
    # Inherits type from super and holds year, make, model, door amount, and roof style.
    
    def __init__(self, type, year=None, make=None, model=None, doors=None, roof=None):
        super().__init__(type)
        self.type = type
        self.year = year
        self.make = make
        self.model = model
        self.doors = doors
        self.roof = roof

    def get_info(self):
        
        # Prompts user for all Automobile attributes.
        
        self.year = input(f"Enter the {self.type}'s year.\n")
        self.make = input(f"Enter the {self.type}'s make.\n")
        self.model = input(f"Enter the {self.type}'s model.\n")
        self.doors = input(f"Enter the number of doors (2 or 4).\n")
        self.roof = input(f"Enter the {self.type}'s roof (solid or sun roof).\n")

    def display_info(self):
        
        # Prints all Automobile attributes.
        
        print(f"\n{'Vehicle type:':<15} {self.type}")
        print(f"{'Year:':<15} {self.year}")
        print(f"{'Make:':<15} {self.make}")
        print(f"{'Model:':<15} {self.model}")
        print(f"{'Doors:':<15} {self.doors}")
        print(f"{'Roof Style:':<15} {self.roof}")


def main():
    vehicle_type = input("Enter the type of vehicle (Car, Truck, Boat, Plane, Bike, etc.).\n")
    my_vehicle = Automobile(vehicle_type)
    my_vehicle.get_info()
    my_vehicle.display_info()
    input("\nPress Enter to Close the Program")


if __name__ == "__main__":
    main()
```
